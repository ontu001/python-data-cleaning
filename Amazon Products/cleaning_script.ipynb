```python
import pandas as pd
```

```python
df = pd.read_csv("amazon_product.csv")
```

```python
df.columns
```

Delete redundant columns.

```python
col_to_keep = ['product_title', 'product_price','product_original_price', 'currency', 'product_star_rating','is_best_seller','product_availability','unit_price', 'unit_count']
len(col_to_keep)
```

```python
df = df[col_to_keep]
```

Renaming columns.

```python
df.columns
```

```python
new_col_names = []
for i in df.columns:
    new_col_names.append(i.capitalize())
```

```python
df.columns = new_col_names
```

Dropping duplicates.

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace= True)
```

Remove the NaN values from the dataset.

```python
df.isna().sum()
```

```python
df.drop(columns= ["Product_original_price", 'Product_availability','Unit_price','Unit_count'],inplace= True)
```

```python
df.dropna(inplace= True)
```

Remove the dollar sign and convert the column into integer

```python
df["Product_price"] = df["Product_price"].str.replace("$","")
```

Convert the Is Best Seller into 0 and 1 for machine learning purpose

```python
df["Product_price"] = df["Product_price"].astype(float)
```

```python
df['Is_best_seller'] = df['Is_best_seller'].apply(lambda x : 1 if x == True else 0)
```

```python
df
```

Export the Cleaned data

```python
df.to_csv("Cleaned Amazon Products.csv", index= False)
```
