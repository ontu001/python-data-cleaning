```python
#Import pandas
import pandas as pd
```

```python
#Read the dataset
hotel_bookings = pd.read_csv("hotel_bookings_uncleaned data.csv")
```

```python
#The dataset
hotel_bookings
```

```python

hotel_bookings.info()
```

```python
#Rename column
hotel_bookings.rename({'adults': 'num_adults', 'children':'num_children', 'babies':'num_babies'}, axis= 1, inplace= True)
hotel_bookings
```

```python
#Check null value percentage
hotel_bookings.isnull().sum() * 100/len(hotel_bookings)
```

```python
hotel_bookings['agent'].describe()
```

```python
#fill the missing value for agent column using Median as this column has outlier
hotel_bookings['agent'].fillna(hotel_bookings['agent'].median(), inplace=True)



##we camn also replace the null value by -1
```

```python
hotel_bookings[hotel_bookings['num_children'].isna()]
hotel_bookings[hotel_bookings['num_children'].isna()][['hotel','is_canceled','country']]
hotel_bookings['country'].fillna('unknown',inplace=True)
hotel_bookings.dropna(subset=['num_children'],inplace=True)
```

```python
hotel_bookings.drop(columns=['company'],inplace=True)
```

```python
hotel_bookings.isnull().sum()
```

```python
#Check Data Type of each data column
hotel_bookings.dtypes
```

```python
hotel_bookings = hotel_bookings.astype({'is_canceled': 'boolean','is_repeated_guest':'boolean', 'num_children':'int','agent':'int'})
hotel_bookings.dtypes
```

```python
hotel_bookings['lead_time'].unique()
hotel_bookings['lead_time'].describe()
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800']

hotel_bookings['lead_time_binned'] = pd.cut(hotel_bookings['lead_time'], bins=bins, labels=labels)

hotel_bookings[['lead_time', 'lead_time_binned']]
```
